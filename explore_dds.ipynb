Treat all columns as strings but preserve NaNs.

```python
for column in df.columns:
    nan_cells = df[column].isna()
    df[column] = df[column].astype(str)
    df.loc[nan_cells, column] = np.nan
df.isna().sum()
```

We will handle the NaNs when we merge the Disaster Declaration Summaries with the Mission Assignments.

Use scatter plot visualizations to learn more about correlations between columns

```python
all_states = (df['state'].unique())
all_states_sorted = sorted(df['state'].unique())
regions = list(range(1,11))

fig = px.scatter(df,
                 x = 'state',
                 y = 'region',
                 color = 'region',
                 title = 'States by Region',
                 labels = {'state' : 'State',
                           'region' : 'Region'},
    )

fig.update_xaxes(
    categoryorder = 'array',
    categoryarray = all_states,
    tickmode = 'linear',
    dtick = 1
)

fig.update_yaxes(
    categoryorder = 'array',
    categoryarray = regions,
    tickmode = 'linear',
    dtick = 1
)

fig.update_layout(
    width = 800 
)

fig.show()
```

```python
fig = px.scatter(df.groupby(['incidentType', 'region']).size().reset_index(name='count'),
                 x = 'incidentType',
                 y = 'region',
                 size = 'count',
                 title = 'Incident Types by Region with Size by Count',
                 labels = {'incidentType' : 'Incident Type',
                           'region' : 'Region'},
    )

fig.update_yaxes(
    categoryorder = 'array',
    categoryarray = regions,
    tickmode = 'linear',
    dtick = 1,
)

fig.update_layout(
    width = 800,
    height = 600,
)

fig.show()
```

```python
fig = px.scatter(df.groupby(
        ['incidentType', 'state']).size(       
    ).reset_index(
        name='count').sort_values(by='state'),
                 x = 'incidentType',
                 y = 'state',
                 size = 'count',
                 title = 'Incident Types by State with Size by Count',
                 labels = {'incidentType' : 'Incident Type',
                           'state' : 'State'},
    )

fig.update_yaxes(
    categoryorder = 'array',
    categoryarray = all_states,
    tickmode = 'linear',
    dtick = 1
)

fig.update_layout(
    height = 1000,
    width = 800,
)

fig.show()
```

```python
fig = px.scatter(df.groupby(
        ['incidentType', 'fyDeclared']).size(       
    ).reset_index(
        name='count').sort_values(by='fyDeclared'),
                 x = 'incidentType',
                 y = 'fyDeclared',
                 size = 'count',
                 title = 'Incident Types by Fiscal Year with Size by Count',
                 labels = {'incidentType' : 'Incident Type',
                           'fyDeclared' : 'Fiscal Year'},
    )

fig.update_yaxes(
    categoryorder = 'array',
    categoryarray = sorted(df['fyDeclared'].unique()),
    tickmode = 'linear',
    dtick = 1,
)

fig.update_layout(
    height = 1200,
    width = 600,
)

fig.show()
```
